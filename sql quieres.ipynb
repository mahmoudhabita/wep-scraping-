# 14) Write a sql query to join both books & categories table into one new table containing book_name & category_name & book_rate and book_price.

```python
cursor.execute('''CREATE TABLE IF NOT EXISTS all_info AS SELECT title AS book_title ,price AS book_price ,rates AS book_rate ,category AS book_category FROM books INNER JOIN catigories ON books.id=catigories.id''')
```

```python
df=pd.read_sql('''select * from all_info''',connection)
df
```
